# 데이터불러오기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
smap = pd.read_excel('/content/drive/MyDrive/과제용.xlsx')
smap.head()
```

```python
%matplotlib inline

import matplotlib as mpl  # 기본 설정
import matplotlib.pyplot as plt  # 그래프 그리기
import matplotlib.font_manager as fm  # 폰트 관리


!apt-get update -qq         # apt-get 패키지 설치 명령어, -qq : 에러외 메세지 숨기기
!apt-get install fonts-nanum* -qq #나눔글꼴 설치

fe = fm.FontEntry(fname=r'/usr/share/fonts/truetype/nanum/NanumGothic.ttf', name='NanumGothic') #파일 저장되어있는 경로와 이름 설정
fm.fontManager.ttflist.insert(0, fe)  # Matplotlib에 폰트 추가
plt.rcParams.update({'font.size': 10, 'font.family': 'NanumGothic'}) #폰트설정
```

```python
smap.info() #인수중량, 잔량, 재단시간 문자형형
```

```python
smap['평량'] = smap['평량'].astype('float64')
smap['가로'] = smap['가로'].astype('float64')
smap['세로'] = smap['세로'].astype('float64')
smap['인수수량'] = smap['인수수량'].astype('float64')
smap['사용중량'] = smap['사용중량'].astype('float64')
smap['수량'] = smap['수량'].astype('float64')

smap['양품속수'] = smap['양품속수'].astype('float64')
smap['불량속수'] = smap['불량속수'].astype('float64')
smap['찍힘수량'] = smap['찍힘수량'].astype('float64')
smap['속도'] = smap['속도'].astype('float64')
smap['재단시간'] = smap['재단시간'].astype('float64')
```

```python
smap['완정지폭'] = smap['완정지폭'].apply(pd.to_numeric, errors = 'coerce')
smap['인수중량'] = smap['인수중량'].apply(pd.to_numeric, errors = 'coerce')
smap['잔량'] = smap['잔량'].apply(pd.to_numeric, errors = 'coerce')
```

```python
plt.figure(figsize=(20,10))
correlation_smap = smap.corr()
sns.heatmap(correlation_smap, annot=True)
```

# 손율상관분석

```python
smapa = smap.copy()
```

```python
smapa.head()
```

```python
smapa = smapa.drop('지종', axis=1)
```

```python
smapa = smapa.drop('반', axis=1)
smapa = smapa.drop('Lot번호', axis = 1)
smapa = smapa.drop('색상', axis = 1)
smapa = smapa.drop('광택', axis = 1)
smapa = smapa.drop('수출', axis = 1)
smapa = smapa.drop('Unnamed: 4', axis = 1)
```

```python
#손율 변수 만들기
smapa['손율'] = (smapa['파지량'])/(smap['재단중량']+ smap['파지량'])
smapa.head()
```

#가동률상관분석

```python
smapb = smap.copy()
```

```python
smapb = smapb.drop('지종', axis = 1)
smapb = smapb.drop('반', axis=1)
smapb = smapb.drop('Lot번호', axis = 1)
smapb = smapb.drop('색상', axis = 1)
smapb = smapb.drop('광택', axis = 1)
smapb = smapb.drop('수출', axis = 1)

smapb = smapb.drop('Unnamed: 4', axis = 1)
```
